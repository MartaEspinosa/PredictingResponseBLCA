```python
import pandas as pd
import numpy as np
import os

GENERAL="/users/genomics/marta/BLCA"
GENOMEDIR="/genomics/users/marta/genomes"
biomart=pd.read_csv(os.path.join(GENOMEDIR,"transcript_gene_v41.txt"), skiprows=1, names=['gene_id','gene_name', 'gene_type'])
biomart_old=pd.read_csv(os.path.join(GENOMEDIR,"transcript_gene.txt"), skiprows=1, names=['gene_id','transcript_id','gene_name_OLD', 'gene_type'])
biomart_old.drop("transcript_id", axis=1, inplace=True)
```

```python
raw_candidates = pd.read_csv(os.path.join(GENERAL,"RNASeq_expression/tumorspecific_3TPM_cohorts.csv"))
raw_candidates = raw_candidates.merge(biomart, on=['gene_type','gene_name'])
candidates = raw_candidates
```

## Ouspenskaia

```python
ouspenskaia = pd.read_csv("/genomics/users/marta/nuORFdb/GSE143263_nuORFdb_v1.0_annotations.csv")
ouspenskaia['gene_id'] = ouspenskaia['geneId'].str.split(".", expand=True)[0]
ouspenskaia_genes = set(ouspenskaia.gene_id.values.tolist())

candidates['nuORF'] = np.where(candidates["gene_id"].isin(ouspenskaia_genes), True, False)
candidates
```

## Chong

```python
chong = pd.read_csv("/genomics/users/marta/Chongetal/Chong_etal_2020_SupData3_41467_2020_14968_MOESM5_ESM.csv", skiprows=1)
chong.rename(columns={'Gene_ID' : 'gene_id'}, inplace=True)
chong_genes = set(chong.gene_id.values.tolist())

candidates['Chong'] = np.where(candidates["gene_id"].isin(chong_genes), True, False)
candidates
```

## vanHeesch

```python
vanHeesch = pd.read_csv("/datasets/marta/vanHeesch2023/SuppTable1.csv", skiprows=1)
vanHeesch.rename(columns={'Gene stable ID' : 'gene_id'}, inplace=True)
vanHeesch_genes = set(vanHeesch.gene_id.values.tolist())

candidates['vanHeesch'] = np.where(candidates["gene_id"].isin(vanHeesch_genes), True, False)
candidates
```

## IEAtlas

```python
epitopes_IEAtlas = pd.read_csv("/genomics/users/marta/IEAtlas/epitopes_in_cancer_tissues_IEAtlas.txt", sep="\t")
print(len(epitopes_IEAtlas))
basic_info_IEAtlas = pd.read_csv("/genomics/users/marta/IEAtlas/epitopes_basic_information_IEAtlas.txt", sep="\t")

IEAtlas = epitopes_IEAtlas.merge(basic_info_IEAtlas, on=["Sequence", "Length", "ORF_ID"])
IEAtlas_genes = set(IEAtlas.geneId.values.tolist())

candidates['IEAtlas'] = np.where(candidates["gene_id"].isin(IEAtlas_genes), True, False)
candidates
```

## Present in HCC and predicted SB ?

```python
projects=['liver_adjacent_totalRNA',"hcc_normal_totalRNA","GSE193567","LIHC_TCGA"]
```

```python
SBs_all = pd.DataFrame(columns=['transcript_id','SB','RiboSeq'])
for proj in projects:
    SBs = pd.read_csv(os.path.join("/projects_eg/projects/marta",proj,"analysis/11_netMHCpan_patientspecific/TPM05/all_potential_epitopes_noredundant_allinformation_SB_RiboSeq_pcandmutexpr_TRUE.csv"))
    SBs = SBs[['ID','SB','RiboSeq']]
    SBs.rename(columns={'ID':'transcript_id'}, inplace=True)
    SBs_all = pd.concat([SBs_all, SBs])
    # transcripts.extend(SBs.ID.values.tolist())

# transcripts = set(transcripts)
# transcripts_df = pd.DataFrame(transcripts, columns=['transcript_id'])
# transcripts_df
SBs_all.drop_duplicates(inplace=True)

# if repeated and different value for SB, keep True
only_SB = SBs_all.drop('RiboSeq', axis=1).drop_duplicates()
only_SB = only_SB.sort_values(by='SB', ascending=False)
only_SB = only_SB.drop_duplicates(subset='transcript_id', keep='first')

# if repeated and different value for SB, keep True
only_RiboSeq = SBs_all.drop('SB', axis=1).drop_duplicates()
only_RiboSeq = only_RiboSeq.sort_values(by='RiboSeq', ascending=False)
only_RiboSeq = only_RiboSeq.drop_duplicates(subset='transcript_id', keep='first')


SBs = only_SB.merge(only_RiboSeq, on="transcript_id")
SBs.drop_duplicates(inplace=True)
SBs
```

```python
biomart_basic = pd.read_csv("/genomics/users/marta/genomes/transcript_gene.txt")
biomart_basic.columns=['gene_id','transcript_id','gene_name','gene_type']

biomart_basic = biomart_basic[['gene_id','transcript_id']]
```

```python
transcripts_genes_SB = SBs.merge(biomart_basic, on="transcript_id")
transcripts_genes_SB.drop("transcript_id", axis=1, inplace=True)
transcripts_genes_SB.columns = ['HCC_predicted_SB','HCC_RiboSeq','gene_id']
```

```python
candidates
```

```python
candidates_HCC = candidates.merge(transcripts_genes_SB, on="gene_id", how="left")
candidates_HCC
```

```python
candidates.to_csv("/users/genomics/marta/BLCA/RNASeq_expression/tumorspecific_3TPM_cohorts_10percent_immuno.csv", index=False)
candidates_HCC.to_csv("/users/genomics/marta/BLCA/RNASeq_expression/tumorspecific_3TPM_cohorts_10percent_riboseq.csv", index=False)
```

```python
candidates_subset = candidates.drop(['UC-GENOME','IMvigor210',"SNY-2017","UNC-108","HdM-BLCA-1","total_n","total_percentage","Chong","vanHeesch"], axis=1)
candidates_subset
# Create a new column based on the specified conditions
candidates_subset['Immunoevidence_Source'] = 'False'  # Initialize the new column with 'False'

# Apply conditions to update the new column
candidates_subset.loc[(candidates_subset['nuORF'] & candidates_subset['IEAtlas']), 'Immunoevidence_Source'] = 'nuORFdb&IEAtlas'
candidates_subset.loc[(candidates_subset['nuORF']) & (~candidates_subset['IEAtlas']), 'Immunoevidence_Source'] = 'nuORFdb'
candidates_subset.loc[(~candidates_subset['nuORF']) & (candidates_subset['IEAtlas']), 'Immunoevidence_Source'] = 'IEAtlas'
```

```python
candidates_subset_biomartOLD = candidates_subset.merge(biomart_old, on=["gene_id","gene_type"], how="left")
candidates_subset_biomartOLD.drop_duplicates(inplace=True)
candidates_subset_biomartOLD
candidates_subset_biomartOLD.drop(["nuORF","IEAtlas","gene_id"], axis=1).to_csv("/users/genomics/marta/BLCA/RNASeq_expression/tumorspecific_3TPM_cohorts_10percent_immunoevicenceSource_geneNames.csv", index=False)
```
